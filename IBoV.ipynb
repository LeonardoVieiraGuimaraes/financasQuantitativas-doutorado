```r
B <-read.csv('IBOV.csv')
head(B)
```

```r
B19 = B[which(B[,1] > 20190000 & B[,1] < 20191231)[1]:nrow(B),]

which(B[,1] > 20190000 & B[,1] < 20191231)[1]:nrow(B)
```

```r
B20 = B[which(B[,1] >= 20200000)[1]:nrow(B),]
```

```r
B19$Date = as.Date(strptime(B19$Date, "%Y%m%d"))
B20$Date = as.Date(strptime(B20$Date, "%Y%m%d"))
```

```r


r19 <- matrix(rep(0,(nrow(B19)-1)*(ncol(B19)-1)), (nrow(B19)-1), (ncol(B19)-1))
for (i in 2:ncol(B19)){
    for (j in 2:nrow(B19)){
        r19[j-1,i-1] <- (B19[j,i] - B19[j-1,i])/B19[j-1,i]
    }
}
```

```r
ncol(r19)
nrow(r19)
nrow(B19)
```

```r
a = cor(t(r19))
a[1,]
```

```r
m19 = rep(0, ncol(B19)-1)
for (i in 2:ncol(B19)) {
    m19[i-1] = mean(B19[,i])
    }
c19 = rep(0, ncol(B19)-1)
for (i in 2:ncol(B19)) {
    
    
    c19 = ((B20[,2] - mean(B19[,2]))*(B19[,i] - mean(B19[,i])))/sqrt(var(B19[,2])*var(B19[,i]))
    c19[i-1] = cor(B19[,2],B19[,i])
    }
```

```r
m20 = rep(0, ncol(B20)-1)
for (i in 2:ncol(B20)) {
    
    m20[i-1] = mean(B20[,i])
    }
c20 = rep(0, ncol(B20)-1)
for (i in 2:ncol(B20)) {
    
    
     
    c20[i-1] = cor(B20[,2],B20[,i])
    }
```
